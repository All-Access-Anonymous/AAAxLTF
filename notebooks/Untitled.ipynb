```python
import random
import numpy as np
from typing import List
```

```python
the_days: List[int] =  [1, 2, 3, 4, 5]
day_distribution: List[float] = [0.1, 0.4, 0.2, 0.2, 0.1]
print([random.choices(the_days, day_distribution) for _ in range(5)])
```
